```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


#convert strings into numerics
school_data["size"].astype(int)
school_data["budget"].astype(int)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])




school_data_complete.head()
```

```python
#create school name series
school_name_series=school_data.loc[:,"school_name"]
```

```python
#create school type Series
school_type_series= school_data.loc[:, "type"]
```

```python
#create total students series
school_students_series= school_data.loc[:, "size"]
```

```python
#create total budget series
school_budget_series= school_data.loc[:, "budget"]
```

```python
#create budget per student series
budget_perstudent_series= school_budget_series/school_students_series
```

```python
# create "Average Math Score" series
avg_math_series= school_data_complete.groupby("school_name", as_index=False).agg({"math_score":"mean"})["math_score"]
```

```python
#create Average Reading Score series
avg_read_series= school_data_complete.groupby("school_name",as_index=False).agg({"reading_score":"mean"})["reading_score"]
```

```python
# create %Passing Math series
passing_math=school_data_complete.loc[school_data_complete['math_score']>=70, ["school_name","math_score"]].groupby("school_name")["math_score"].count()/school_data_complete.groupby("school_name")["math_score"].count()*100
passing_math = passing_math.reset_index()
```

```python
#create % Passing Reading series
passing_reading=school_data_complete.loc[school_data_complete['reading_score']>=70, ["school_name","reading_score"]].groupby("school_name")["reading_score"].count()/school_data_complete.groupby("school_name")["reading_score"].count()*100
passing_reading=passing_reading.reset_index()
```

```python
#create Overall Passing Rate series
overall_passing=(passing_math["math_score"]+passing_reading["reading_score"])/2
overall_passing=overall_passing.reset_index()
```

```python
#populate dataframe
school_summary_df= pd.DataFrame(
    {
    "School Name":school_name_series,
    "School Type":school_type_series,
    "Total Students":school_students_series,
    "Total School Budget":school_budget_series,
    "Per Student Budget":budget_perstudent_series,
    "Average Math Score":avg_math_series,
    "Average Reading Score":avg_read_series,
    "% Passing Math":passing_math.iloc[:,1],
        "% Passing Reading":passing_reading.iloc[:,1],
    "Overall Passing Rate":overall_passing.iloc[:,1]
    }
    )
school_summary_df.head(20)
```
